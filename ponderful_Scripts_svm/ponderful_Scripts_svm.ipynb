```python
import pandas as pd
from sklearn import svm
import sklearn.model_selection as model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
#Data reading 
Phy_Che = pd.read_pickle('/Users/lidiayung/PhD_project/project_PONDERFUL/ponderful_OUTPUT/Phy_Che_qt.pkl')
display(Phy_Che)
```

```python
Phy_Che['eco_regions']='Temperate'
Phy_Che.loc[Phy_Che['Country'].str.contains('Uruguay'), 'eco_regions'] = 'Subtropical'
Phy_Che.loc[Phy_Che['Country'].str.contains('Turkey'), 'eco_regions'] = 'Continental'
Phy_Che.loc[Phy_Che['Country'].str.contains('Spain'), 'eco_regions'] = 'Mediterranean'
```

```python
vars=['Pond_ID', 'PondCode', 'Pondscape', 'Country', 'Strat_Survey',
       'Resampling', 'season_col','Area.t', 'Depth.t','TN', 'TP','bio1.t', 'bio4.t', 'bio5.t','bio12.t','Hydeoperiod_length.t',
       'Animals_cont.t', 'Natural_5.t','Aquatic_500_qt', 'Cropland_500_qt', 'Forest_500_qt',
       'Pastures.and.open.nature_500_qt','Urban_500_qt','eco_regions']
Phy_Che_rf_df=Phy_Che[vars]
```

```python
Phy_Che_rf_df=Phy_Che_rf_df.dropna()
```

```python
# Split the data into features (X) and target (y)
X = Phy_Che_rf_df[['Area.t', 'Depth.t','bio1.t', 'bio4.t', 'bio5.t','bio12.t','Hydeoperiod_length.t',
'Animals_cont.t', 'Natural_5.t','Aquatic_500_qt', 'Cropland_500_qt', 'Forest_500_qt',
'Pastures.and.open.nature_500_qt','Urban_500_qt']]

y = Phy_Che_rf_df['eco_regions']


# Split the data into training and test sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.80, test_size=0.20, random_state=101)
```

```python
rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1).fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1).fit(X_train, y_train)
```

```python
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
```

```python
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
```

```python
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
print('Accuracy (RBF Kernel): ', "%.2f" % (rbf_accuracy*100))
print('F1 (RBF Kernel): ', "%.2f" % (rbf_f1*100))
```
